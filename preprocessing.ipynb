```python
# Import requisite libraries and upload file
import pandas as pd
from google.colab import files
uploaded = files.upload()
dataset = pd.read_csv('Dataset.csv')
datasetwithNaN = dataset
dataset
```

```python
# To get the data summary

print(dataset.shape)                   # To get the dataset number of rows and columns
print(dataset.describe())              # To get the dataset summary: data mean, std, count, min, 25%, 50% and 75%
dataset.isna().any()                   # To check for NaN values. Run and we'll get land_false (means no empty row) and price_true means there's empty row

```

```python
# Replace the NaN value with Mean
MeandatasetNotNaN = dataset.copy()                                                   # Creates a new dataframe copy so that the original dataset is not altered.
MeandatasetNotNaN['price'] = dataset['price'].fillna(dataset['price'].mean())        # Assigns the price column with NaNs filled by the mean of price.
print(MeandatasetNotNaN)                                                             # Print with the NaN replaced by mean.
```

```python
# Replace the NaN value with Median because the presence of the outlier causes the mean to be skewed.
MediandatasetNotNaN = dataset.copy()
MediandatasetNotNaN['price'] = dataset['price'].fillna(dataset['price'].mean())
print(MediandatasetNotNaN)
```

```python
# Calculate the Q1 (25th percentile) and Q3 (75th percentile) for the price column
Q1 = dataset['price'].quantile(0.25)
Q3 = dataset['price'].quantile(0.75)
IQR = Q3-Q1


# Define the lower and upper bounds for outlier
lower_bound = Q1 -1.5 *IQR
upper_bound = Q3 +1.5 *IQR

# Remove outliers from the dataset
dataset_no_outliers = dataset [(dataset['price']>=lower_bound) & (dataset['price']<=upper_bound)]
print(dataset_no_outliers)


```

```python
# Option Two: To replace the outlier. (Simplistic Approach)

percentile = dataset.price.quantile(1.0)                  # Looks for the max
percentile

datasetNoOutlier = dataset[dataset.price<percentile]      # Removes the outlier
datasetNoOutlier
```
